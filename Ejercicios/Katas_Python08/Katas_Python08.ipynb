## üìò Katas de Python 08 - Divisi√≥n con manejo de excepciones

Escribe un programa que pida al usuario dos n√∫meros e intente dividirlos. Si el usuario ingresa un valor no num√©rico 
o intenta dividir por cero, maneja esas excepciones de manera adecuada. Aseg√∫rate de mostrar un mensaje 
indicando si la divisi√≥n fue exitosa o no

```python
# Katas de Python 08 - Divisi√≥n segura con manejo de errores

def dividir_usuario():
    """
    Pide al usuario dos n√∫meros e intenta dividirlos de forma segura.
    
    paso 1: Pedir al usuario el primer n√∫mero
    paso 2: Pedir al usuario el segundo n√∫mero
    paso 3: Intentar realizar la divisi√≥n dentro de un bloque try/except
    paso 4: Capturar ValueError si el usuario no introduce un n√∫mero
    paso 5: Capturar ZeroDivisionError si el segundo n√∫mero es cero
    paso 6: Mostrar un mensaje indicando si la divisi√≥n fue exitosa o no
    paso 7: Devolver el resultado si la divisi√≥n se realiz√≥ correctamente
    """
    try:
        # 1Ô∏è‚É£ Pedir el primer n√∫mero
        num1 = float(input("introduce el primer numero:"))

        # 2Ô∏è‚É£ Pedir el segundo n√∫mero
        num2 = float(input("introduce el segundo numero:"))

        # 3Ô∏è‚É£ Intentar realizar la divisi√≥n
        resultado = num1 / num2

        # 6Ô∏è‚É£ Mostrar mensaje de √©xito
        print(f"division exitosa: {num1} / {num2} = {resultado}")

        # 7Ô∏è‚É£ Devolver el resultado
        return round(resultado, 2)
    
    except ValueError:
        # 4Ô∏è‚É£ Mensaje de error si no es un n√∫mero
        print("Error: Debes introducir n√∫meros v√°lidos.")   
    except ZeroDivisionError:
        # 5Ô∏è‚É£ Mensaje de error si el segundo n√∫mero es cero
        print("Error: No se puede dividir por cero.")
```

```python
# ejemplos de uso 

dividir_usuario()
```

## üí° BONUS - registro de datos

```python
def generar_dataset_divisiones():
    """
    Ejecuta dividir_usuario() repetidamente hasta que el usuario decida parar, 
    guardando los resultados v√°lidos en una lista.
    
    paso 1: Inicializar una lista vac√≠a para almacenar los resultados v√°lidos
    paso 2: Iniciar un bucle infinito (while True) para la captura de datos
    paso 3: Llamar a la funci√≥n 'dividir_usuario()' para obtener un dato
    paso 4: Si el dato es v√°lido (no es None), agregarlo a la lista
    paso 5: Preguntar al usuario si desea detener el proceso (salir)
    paso 6: Si el usuario confirma la salida, romper el bucle
    paso 7: Mostrar la lista final con todos los registros acumulados
    """
    
    # 1Ô∏è‚É£ Inicializar lista para almacenar resultados
    dataset_resultados = []
    
    # 2Ô∏è‚É£ Iniciar bucle de captura
    while True:
        print("\n--- Nuevo Ingreso ---")
        
        # 3Ô∏è‚É£ Llamar a la funci√≥n dividir_usuario
        dato = dividir_usuario()
        
        # 4Ô∏è‚É£ Agregar a la lista si es v√°lido
        if dato is not None:
            dataset_resultados.append(dato)
            
        # 5Ô∏è‚É£ Preguntar si desea detener el proceso
        detener = input("¬øQuieres terminar el proceso? (si/no): ")
        
        # 6Ô∏è‚É£ Romper el bucle si el usuario dice 'si'
        if detener.lower() == 'si':
            break
            
    # 7Ô∏è‚É£ Mostrar la lista final (Tu Dataset)
    print("\n" + "="*30)
    print(f"DATASET FINAL: {dataset_resultados}")
    print("="*30)


```

```python
# Ejecutar el bonus
generar_dataset_divisiones()
```
