## üìò Katas de Python 03 - Buscar palabras que contienen un objetivo

Escribe una funci√≥n que tome una lista de palabras y una palabra objetivo como par√°metros. La funci√≥n debe 
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
# katas de Python 03 - Buscar palabras que contienen un objetivo

def buscar_palabras(lista_palabras, objetivo):
    """
    Devuelve una nueva lista con las palabras que contienen la palabra 'objetivo'.

    paso 1 : Crear una lista vac√≠a para guardar coincidencias
    paso 2 : Recorrer cada palabra de la lista original
    paso 3 : Comprobar si 'objetivo' est√° dentro de la palabra
    paso 4 : Si contiene el objetivo, agregarla a la lista
    paso 5 : Devolver la lista final
    """

    coincidencias = []  # 1Ô∏è‚É£ Lista vac√≠a

    for palabra in lista_palabras:  # 2Ô∏è‚É£ Recorrer lista
        if objetivo in palabra:     # 3Ô∏è‚É£ Comprobar coincidencia
            coincidencias.append(palabra)  # 4Ô∏è‚É£ Agregar coincidencia

    return coincidencias  # 5Ô∏è‚É£ Devolver resultado
```

```python
# ejemplos de uso

lista = ["casa", "caseta", "perro", "casco", "cami√≥n", "cascada"]
objetivo = "cas"

print(buscar_palabras(lista, objetivo))
```

```python
# tests autom√°ticos para verificar que la funci√≥n funciona correctamente

assert buscar_palabras(["hola", "holanda", "adios"], "hola") == ["hola", "holanda"]
```

## üí° BONUS - Contar cu√°ntas palabras coinciden

Vamos a crear la misma version , pero vamos a contar cu√°ntas palabras coinciden.

```python
#bonus 3: contar palabras que contienen un objetivo

def contar_palabras_coincidentes(lista_palabras, objetivo):
    """
    Devuelve dos cosas:
      1. Una lista con las palabras que contienen 'objetivo'
      2. La cantidad TOTAL de coincidencias

    paso 1 : Convertir la palabra objetivo a min√∫sculas
    paso 2 : Crear una lista vac√≠a para guardar coincidencias
    paso 3 : Recorrer la lista de palabras
        - Convertir cada palabra a min√∫sculas
        - Si contiene la palabra objetivo, agregarla a la lista
    paso 4 : Calcular la cantidad de coincidencias
    paso 5 : Devolver la lista y la cantidad
    """

    objetivo = objetivo.lower()  # 1Ô∏è‚É£ Normalizar objetivo
    coincidencias = []    # 2Ô∏è‚É£ Lista vac√≠a

    for palabra in lista_palabras: # 3Ô∏è‚É£ Recorrer palabras
        if objetivo in palabra.lower():
            coincidencias.append(palabra)

    cantidad = len(coincidencias) # 4Ô∏è‚É£ Calcular cantidad

    return coincidencias, cantidad # 5Ô∏è‚É£ Devolver lista y cantidad
```

```python
# ejemplos de uso

palabras = ["computador", "composici√≥n", "compra", "naranja", "COMPUTAR"]
resultado, cantidad = contar_palabras_coincidentes(palabras, "com")

print("Coincidencias:", resultado)
print("Cantidad total de coincidencias:", cantidad)
```
