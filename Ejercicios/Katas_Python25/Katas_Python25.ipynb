## ðŸ“˜ Katas de Python 25 - Contador de Caracteres

Crea una funciÃ³n que cuente el nÃºmero de caracteres en una cadena de texto dada.

```python
# Katas de Python 25 - Contador de Caracteres

def contar_caractares(texto):
    """
    Recibe una cadena de texto y devuelve cuÃ¡ntos caracteres tiene 
    (incluyendo espacios y sÃ­mbolos).
    
    paso 1: Usar la funciÃ³n len() sobre el texto
    paso 2: Devolver el resultado
    """

    # paso 1: utilizamos la funcion len()
    cantidad = len(texto)

    # paso 2: devuelve resultado
    return cantidad
```

```python
# ejemplos de uso

mensaje = "Soy david Morales Mendez"
resultado = contar_caractares(mensaje)

print(f"Texto: '{mensaje}'")
print(f"Longitud: {resultado} caracteres")
```

## ðŸ’¡ BONUS - CreaciÃ³n de ContraseÃ±a (Interactivo)


```python
# BONUS - CreaciÃ³n de ContraseÃ±a (Interactivo)


def pedir_password():
    """
    Obliga al usuario a introducir una contraseÃ±a vÃ¡lida (entre 4 y 10 caracteres).
    Si no cumple, vuelve a preguntar infinitamente.
    """
    
    print("--- ðŸ” CONFIGURACIÃ“N DE SEGURIDAD ---")
    print("Reglas: MÃ­nimo 4 caracteres | MÃ¡ximo 10 caracteres")
    
    # paso 2: BUCLE INFINITO (Se repite una y otra vez)
    while True:
        
        # Pedimos el dato dentro del bucle
        password = input("\nIntroduce nueva contraseÃ±a: ")
        
        # Medimos con TU funciÃ³n
        longitud = contar_caractares(password)
        
        # paso 3: VALIDACIONES QUE OBLIGAN A REPETIR ---
        
        if longitud > 10:
            print(f"â›” ERROR: {longitud} es demasiado largo (MÃ¡x 10). EscrÃ­bela de nuevo.")
            # Al no poner 'break', el bucle vuelve arriba y pregunta otra vez
            
        elif longitud < 4:
            print(f"â›” ERROR: {longitud} es demasiado corto (MÃ­n 4). EscrÃ­bela de nuevo.")
            # El bucle vuelve a empezar
            
        # paso 4: CASOS VÃLIDOS (ROMPE EL BUCLE) ---
        
        else:
            # Si llegamos aquÃ­, estÃ¡ entre 4 y 10. Analizamos la calidad:
            if longitud == 6:
                print("âš ï¸ AVISO: Es vÃ¡lida, pero un poco INSEGURA.")
            elif longitud == 8:
                print("âœ… PERFECTO: Es una contraseÃ±a SEGURA.")
            else:
                print("ðŸ‘ ContraseÃ±a guardada.")
            
            # El break rompe la cÃ¡rcel del while y termina el programa
            break
```

```python
pedir_password()
```
