##  Katas de Python 22 - Producto Total

Dada una lista num茅rica, obt茅n el producto total de los valores de dicha lista.Usa la funci贸n 
reduce()

```python
# Katas de Python 22 - Producto Total

# importamos la funcion reduce
from functools import reduce
 
def obtener_producto_total(lista_numeros):
    """
    Multiplica todos los valores de una lista usando reduce.

    paso 1: Definir una funci贸n que sepa multiplicar dos n煤meros
    paso 2: Usar reduce() para aplicar esa funci贸n a toda la lista
    """

    # paso 1: funcion para multiplicar dos numeros
    def multiplicar(a, b):
        return a * b
    
    # paso 2: reduce itera con la lista y los multiplica
    resultado = reduce(multiplicar, lista_numeros)

    return resultado
```

```python
# ejemplos de uso 

mis_numeros =[3, 4, 5, 6]

total = obtener_producto_total(mis_numeros)

print(f"La lista original es:{mis_numeros}")
print(f"el resultado total es:{total}")
```

##  BONUS - Comparador de resultados (A vs B)

```python
# BONUS - Comparador de resultados (A vs B)

def comparar_resultados(lista_A, lista_B):
    """
    Recibe dos listas de multiplicadores, calcula el total de cada una
    usando la funci贸n anterior , y dice cu谩l es la ganadora.
    
    paso 1: Calcular el total de la Lista A 
    paso 2: Calcular el total de la Lista B 
    paso 3: Comparar los resultados y dice cual es el ganador
    """

    # paso 1: calculamos el total A
    total_A = obtener_producto_total(lista_A)
    print(f"la lista a es: {lista_A} ----> el total es: {total_A}")

    # paso 2: calculamos el total B
    total_B = obtener_producto_total(lista_B)
    print(f"la lista b es: {lista_B} ----> el total es: {total_B}")

    # paso 3: comparamos resultados
    if total_A > total_B:
        print(f"gana la Estrategia A (Por {total_A - total_B} puntos)")
    elif total_B > total_A:
        print(f"gana la Estrategia B (Por {total_B - total_A} puntos)")
    else:
        print("el resultado es empate, Ambas dan el mismo resultado.")
```

```python
# ejemplo de uso 

opcion_a = [2, 2, 3, 5]
opcion_b = [5, 2]

comparar_resultados(opcion_a,opcion_b)
```
