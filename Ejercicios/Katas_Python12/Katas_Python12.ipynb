## üìò Katas de Python 12 - Longitud de Palabras

Genera una funci√≥n que al recibir una frase devuelva una lista con la longitud de cada palabra. Usa la funci√≥n map()

```python
# Katas de Python 12 - Longitud de Palabras

def obtener_longitudes(frase):
    """
    Recibe una frase y devuelve una lista con el tama√±o de cada palabra.
    
    paso 1: Dividir la frase en palabras individuales (usando split)
    paso 2: Usar map() para aplicar la funci√≥n len() a cada palabra
    paso 3: Convertir el objeto map resultante en una lista
    paso 4: Devolver la lista de n√∫meros
    """
    # 1Ô∏è‚É£ Convertir texto a lista ("Me llamo David" -> ["Me", "llamo", "David"])
    lista_palabras = frase.split()

    # 2Ô∏è‚É£ Usar map() para obtener la longitud de cada palabra
    longitudes = map(len, lista_palabras)

    # 3Ô∏è‚É£ Convertir el objeto map a lista
    lista_numeros = list(longitudes)

    # 4Ô∏è‚É£ Devolver la lista de n√∫meros
    return lista_numeros
```

```python
# ejemplos de uso

texto = "Me llamo David y tengo 30 a√±os"
resultado = obtener_longitudes(texto)

print(f"Frase: {texto}")
print(f"Longitudes de palabras: {resultado}")
```

## üí° BONUS - Encontrar M√°ximos y M√≠nimos

```python
# BONUS - Encontrar M√°ximos y M√≠nimos

def encontrar_limites(lista_longitudes):
    """
    Recibe la lista de longitudes y muestra el valor m√°s alto y el m√°s bajo.
    Complementa el ejercicio buscando los extremos de los datos.

    paso 1: validamos la lista no est√© vac√≠a
    paso 2: Encontrar el n√∫mero mayor con max()
    paso 3: Encontrar el n√∫mero menor con min()
    paso 4: Mostrar los resultados
    """
    # 1Ô∏è‚É£ Validar que la lista no est√© vac√≠a
    if len(lista_longitudes) > 0:
        # 2Ô∏è‚É£ El dato m√°s alto (Techo)
        mayor = max(lista_longitudes)
        # 3Ô∏è‚É£ El dato m√°s bajo (Suelo)
        menor = min(lista_longitudes)
        # 4Ô∏è‚É£ Mostrar resultados
        print("--- Rango de Datos ---")
        print(f"La palabra m√°s larga tiene: {mayor} letras")
        print(f"La palabra m√°s corta tiene: {menor} letras")
    
    else:
        print("La lista de longitudes est√° vac√≠a.")       
    
```

```python
# ejemplos de uso conjunto

# 1. Obtenemos la lista de longitudes con TU funci√≥n anterior
frase = "Me gusta aprender Python"
mis_numeros = obtener_longitudes(frase)

# 2 el bonus analiza la lista obtenida
encontrar_limites(mis_numeros)
```
