## 游닂 Katas de Python 29 - Enmascarar Datos

Crea una funci칩n que convierta una variable en una cadena de texto y enmascare todos los caracteres  con el 
car치cter '#', excepto los 칰ltimos cuatro.

```python
def enmascarar_datos(dato_secreto):
    """
    Recibe cualquier dato (n칰mero o texto), lo convierte a string
    y oculta todo menos los 칰ltimos 4 caracteres.
    
    paso 1: Convertir el dato a string (str)
    paso 2: a침adimos una seguridad para menores de 4 caracteres
    paso 3: Separar la parte visible (los 칰ltimos 4)
    paso 4: Calcular cu치ntos hay que tapar
    paso 5: Crear la m치scara de almohadillas (#) y unirlo todo
    """

    # paso 1: conversion a str
    texto = str(dato_secreto)
    longitud = len(texto)

    # paso 2: seguridad de menores de 4 digitos
    if longitud <= 4:
        return "丘멆잺 Error: Faltan d칤gitos"

    # paso 3: separamos la parte visible
    # (Si el texto es corto, ej: "123", esto coge "123" entero)
    parte_visible = texto[-4:]

    # paso 4: calculamos cuantos tenemos que enmascarar
    cantidad_a_tapar = longitud - 4
    
    # si la palabra es mas corta, no deberia romper el codigo
    # (Python maneja multiplicaciones negativas devolviendo vac칤o, pero esto lo hace expl칤cito)
    if cantidad_a_tapar < 0:
        cantidad_a_tapar = 0
    
    # paso 5: crear las almohadillas y unirlos
    mascara = "#" * cantidad_a_tapar
    resultado = mascara + parte_visible

    return resultado
```

```python
# ejemplos de uso sin error

tarjeta = 12456325478965
secreto = enmascarar_datos(tarjeta)

print(f"La tarjeta original es: {tarjeta}")
print(f"La tarjeta enmascarada es: {secreto}")
```

```python
# ejemplos de uso con error

tarjeta = 124
secreto = enmascarar_datos(tarjeta)

print(f"La tarjeta original es: {tarjeta}")
print(f"La tarjeta enmascarada es: {secreto}")
```
