## üìò Katas de Python 16 - Filtro de Palabras Largas

 Escribe una funci√≥n que tome una cadena de texto y un n√∫mero entero n como par√°metros y devuelva una lista de 
todas las palabras que sean m√°s largas que n. Usa la funci√≥n 
filter()

```python
# Katas de Python 16 - Filtro de Palabras Largas

def filtrar_palabras_largas(texto,n):
    """
    Recibe un texto y un n√∫mero n. Devuelve las palabras que tengan m√°s de n letras.
    
    paso 1: Convertir el string gigante en una lista de palabras individuales (split)
    paso 2: Definir la condici√≥n: len(palabra) > n
    paso 3: Usar filter() para aplicar esa condici√≥n a la lista
    paso 4: Devolver la lista de palabras filtradas
    """
    # 1Ô∏è‚É£ Trocear el texto (rompe por los espacios)
    lista_palabras = texto.split()

    # 2Ô∏è‚É£ Funci√≥n auxiliar para contar caracteres
    def es_larga(palabra):
        return len(palabra) > n
    
    # 3Ô∏è‚É£ Aplicamos el Filtro
    # solo pasan las palabras que cumple la condicion
    resultado = filter(es_larga, lista_palabras)

    # 4Ô∏è‚É£ Retorno y pasamos a lista
    return list(resultado)
```

```python
# ejemplo de uso 

# 1. a√±adimos los datos 
frase = "el partido de futbol estuvo increiblemente bueno y atractivo"
minimo_letras = 5

# 2. llamamos a nuestra funcion filtrar
resultado = filtrar_palabras_largas(frase, minimo_letras)

# 3. obtenemos los resultados
print(f"texto original: {frase}")
print(f"las palabras de mas de {minimo_letras} letras son : {resultado}")
```

## üí° BONUS - Generador de Hashtags

```python
# BONUS - Generador de Hashtags

def generar_hashtags(texto_redes):
    """
    Crea hashtags a partir de un texto usando tu funci√≥n de filtrado.
    Sin lambdas, solo bucles sencillos.
    
    paso 1: Definir la longitud m√≠nima (ej: 4 letras)
    paso 2: Usar TU funci√≥n 'filtrar_palabras_largas' para limpiar el texto
    paso 3: Recorrer la lista limpia y ponerle '#' a cada palabra
    paso 4: Guardar los hashtags en una lista nueva y mostrarla
    """
    # 1Ô∏è‚É£ Configuraci√≥n
    largo_minimo = 5

    # 2Ô∏è‚É£ Usamos la funcion (filtrar_palabras_largas)
    palabras_clave = filtrar_palabras_largas(texto_redes, largo_minimo)

    # 3Ô∏è‚É£ Formateo donde a√±adimos "#" al hashtag
    lista_hashtags = []

    for palabra in palabras_clave:
        # creamos el hashtag
        etiqueta = f"#{palabra}"
        # guardamos el hashtag
        lista_hashtags.append(etiqueta)
    
    # 4Ô∏è‚É£ Resultado
    print(lista_hashtags)
```

```python
# ejemplo de uso 

# 1. a√±adimos los datos
post = "en cada partido que jugamos, lo hacemos mejor"

# 2. ejecutamos 
generar_hashtags(post)
```
