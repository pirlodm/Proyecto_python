## üìò Katas de Python 14 - Filtrar por Inicial

Crea una funci√≥n que retorne las palabras de una lista de palabras que comience con una letra en especifico. Usa la 
funci√≥n filter()

```python
# Katas de Python 14 - Filtrar por Inicial

def filtrar_por_inicial(lista_palabras, letra_buscada):
    """
    Retorna las palabras que comienzan con la letra indicada, 
    ignorando si est√°n en may√∫sculas o min√∫sculas.
    
    paso 1: Definir funci√≥n auxiliar
    paso 2: Convertir la palabra a min√∫scula (.lower())
    paso 3: Convertir la letra buscada a min√∫scula (.lower())
    paso 4: Comparar usando startswith
    paso 5: Aplicar filter y devolver lista
    """

    # 1Ô∏è‚É£ Funci√≥n Auxiliar(funcion inteligente)
    def empieza_con_letra(palabra):

        # 2Ô∏è‚É£ Normalizamos la palabra del dato (Ana -> ana)
        palabra_limpia = palabra.lower()
        
        # 3Ô∏è‚É£ Normalizamos la letra que buscamos (A -> a)
        letra_limpia = letra_buscada.lower()
        
        # 4Ô∏è‚É£ Comparamos 
        return palabra_limpia.startswith(letra_limpia)
    
    # 5Ô∏è‚É£ Aplicar Filtro
    resultado = filter(empieza_con_letra, lista_palabras)

    return list(resultado)
```

```python
# ejemplos de uso 

nombres = ["Ana", "Bernardo", "Andr√©s", "Carla", "Alberto", "antonio"]
letra = "A"

resultado = filtrar_por_inicial(nombres, letra)

print(f"listas completa: {nombres}")
print(f"filtradas por '{letra}': {resultado}")
```

## üí° BONUS - Bonus: Informe de Resultados

```python
# BONUS - Bonus: Informe de Resultados

def generar_informe(lista_total, lista_filtrada):
    """
    Genera una m√©trica basada en el filtro realizado.
    Nos dice qu√© tan com√∫n es la letra buscada .
    
    paso 1: Contar cu√°ntos datos ten√≠amos al principio (Total)
    paso 2: Contar cu√°ntos datos encontramos con el filtro (Encontrados)
    paso 3: evitar la division entre 0
    paso 4: Calcular el porcentaje: (Encontrados / Total) * 100
    paso 5: Mostrar el informe formateado
    """
    # 1Ô∏è‚É£ y 2Ô∏è‚É£ Obtener las cifras
    total_registros = len(lista_total)
    total_encontrados = len(lista_filtrada)

    # 3Ô∏è‚É£ validacion: evitar division entre 0
    if total_registros == 0:
        print("no hay datos que analizar")
        return
    
    # 4Ô∏è‚É£ calculo del indicador
    porcentaje = (total_encontrados / total_registros) * 100

    # 5Ô∏è‚É£ reporte visual 

    print("-------INFORME DE BUSQUEDA-------")
    print(f"Total Registros: {total_registros}")
    print(f"coincidencias: {total_encontrados}")
    print(f"indicador: {porcentaje} % del total")
```

```python
# ejemplo de uso 

clientes = ["Ana", "alberto", "Beatriz", "carlos", "antonio", "David"]
resultados = filtrar_por_inicial(clientes, "A")
generar_informe(clientes, resultados)
```
