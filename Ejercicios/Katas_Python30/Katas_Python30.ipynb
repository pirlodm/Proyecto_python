## üìò Katas de Python 30 - Detector de Anagramas

Crea una funci√≥n que determine si dos palabras son anagramas, es decir, si est√°n formadas por las mismas letras 
pero en diferente orden

```python
# Katas de Python 30 - Detector de Anagramas

def es_anagrama(palabra1, palabra2):
    """
    Comprueba si dos palabras tienen exactamente las mismas letras.
    
    paso 1: Convertir a min√∫sculas (para evitar errores por may√∫sculas)
    paso 2: Ordenar alfab√©ticamente las letras de ambas palabras
    paso 3: Comparar si las listas ordenadas son id√©nticas
    """

    # paso 1: volvemos todo a minusculas
    p1_limpia = palabra1.lower()
    p2_limpia = palabra2.lower()

    # paso 2: ordenamos las letras alfabeticamente con sorted()
    p1_ordenada = sorted(p1_limpia)
    p2_ordenada = sorted(p2_limpia)

    # paso 3: comparamos las listas ordenadas para ver si son identicas
    if p1_ordenada == p2_ordenada:
        return True
    else:
        return False
    
```

```python
# ejemplos de uso 

# ejemplo 1: son anagramas
print(f"¬ø'Amor' y 'Roma'? -> {es_anagrama('Amor', 'Roma')}")

# ejemplo 2: No son anagramas
print(f"¬ø'Hola' y 'Adios'? -> {es_anagrama('Hola', 'Adios')}")
```

## üí° BONUS - buscar errores mal escritos

```python
# BONUS - buscar errores mal escritos

def buscar_coincidencias_sucias(lisa_clientes, apellido_buscado):
    """
    Busca en una lista si existe el apellido, incluso si tiene las letras bailadas.
    
    paso 1: Recorrer la lista de clientes
    paso 2: Usar la funci√≥n 'es_anagrama' para comparar
    paso 3: Si da True, avisar de que hemos encontrado un posible duplicado
    """
    print(f"---  BUSCANDO VARIANTES DE: '{apellido_buscado}' ---")

    encontrados = []

    # paso 1: recorremos lista cliente
    for cliente in lisa_clientes:
        # paso 2:usamos la funcion  es_anagrama
        if es_anagrama(cliente, apellido_buscado):
            encontrados.append(cliente)
    
    # paso 3: reporte de resultados
    if len(encontrados) > 0:
        print(f"‚ö†Ô∏è Se han encontrado coincidencias:")
        print(f"   {encontrados}")
    else:
        print("‚úÖ No se han encontrado variantes.")
```

```python
# ejemplos de uso 

# Base de datos con errores
base_datos_apellidos = ["Perez", "Gracia", "Lopez", "Garcia", "Agracia"]

# Buscamos "Garcia"
buscar_coincidencias_sucias(base_datos_apellidos, "Garcia")
```
