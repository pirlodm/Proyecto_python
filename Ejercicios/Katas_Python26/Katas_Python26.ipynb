## üìò Katas de Python 26 - Calcular el Resto (Lambda)

Crea una funci√≥n 
lambda que calcule el resto de la divisi√≥n entre dos n√∫meros dados.

```python
# Katas de Python 26 - Calcular el Resto (Lambda)

# definimos lamdba
# recibe dos numeros y devuelve lo que sobra al dividirlos

calcular_resto = lambda a, b: a % b
```

```python
# ejemplos de uso 

numero = 11
divisor = 3

resultado = calcular_resto(numero, divisor)

print(f" dividir {numero} entre {divisor}")
print(f"El resto es: {resultado}")
```

## üí° BONUS - Conversor de Tiempo

```python
def convertir_a_minutos(total_minutos):
    """
    Transforma minutos brutos en un formato legible.
    
    paso 1: Definir el bloque de tiempo (1 hora = 60 minutos)
    paso 2: Calcular las HORAS completas
    paso 3: Calcular los MINUTOS restantes usando la funcion anterior
    paso 4: Mostrar el resultado final formateado
    """
    
    print(f"--- ‚è±Ô∏è TRANSFORMANDO {total_minutos} MINUTOS ---")
    
    # paso 1: Definir la unidad
    bloque_hora = 60
    
    # paso 2: Calcular Horas (La parte entera)
    horas = total_minutos // bloque_hora
    
    # paso 3: Calcular el Resto usando calcular_resto
    minutos_sueltos = calcular_resto(total_minutos, bloque_hora)
    
    # paso 4: Salida de datos (Reporte)
    print(f"Resultado: {horas} horas y {minutos_sueltos} minutos.")

```

```python
# ejemplo de uso

partido_futbol = 180
pelicula = 210

convertir_a_minutos(partido_futbol)
convertir_a_minutos(pelicula)
```
