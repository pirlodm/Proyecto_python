## üìò Katas de Python 05 - Media y estado de aprobado

Ecribe una funci√≥n que tome una lista de n√∫meros como par√°metro y un valor opcional nota_aprobado, que por 
defecto es 5. La funci√≥n debe calcular la media de los n√∫meros en la lista y determinar si la media es mayor o igual 
que nota aprobado. Si es as√≠, el estado ser√° "aprobado", de lo contrario, ser√° "suspenso". La funci√≥n debe devolver 
una tupla que contenga la media y el estado

```python
# Katas de Python 05 - Media y estado de aprobado

def calcular_media_estado(numeros, nota_aprobado=5):
    """
    Calcula la media de una lista de n√∫meros y determina si est√° aprobada.

    paso 1 : Sumar todos los n√∫meros de la lista
    paso 2 : Dividir la suma entre la cantidad de elementos para obtener la media
    paso 3 : Comparar la media con la nota de aprobado
    paso 4 : Devolver una tupla con la media y el estado ("aprobado" o "suspenso")
    """

    # 1Ô∏è‚É£ Sumar los n√∫meros
    suma_total = sum(numeros)

    # 2Ô∏è‚É£ Calcular la media
    media = suma_total / len(numeros)

    # 3Ô∏è‚É£ Determinar el estado
    if media >= nota_aprobado:
        estado = "est√° aprobado"
    else:
        estado = "est√° suspenso"

    # 4Ô∏è‚É£ Devolver la tupla, (media y estado)
    return media, estado
```

```python
# ejemplos de uso 

lista_notas = [6, 7, 8, 5, 9]
print(calcular_media_estado(lista_notas))
print(calcular_media_estado(lista_notas, nota_aprobado=8))  # con nota de aprobado personalizada
```

## üí° BONUS - Contar aprobados 

```python
# Katas de Python 05 - Media y estado (con bonus: contar aprobados)

def calcular_media_estado(numeros, nota_aprobado=5, bonus=False):
    """
    Calcula la media de una lista de notas y determina si la media est√° aprobada o no.

    paso 1 : Verificar que la lista no est√© vac√≠a
    paso 2 : Sumar todos los valores y dividir por la cantidad para obtener la media
    paso 3 : Comparar la media con 'nota_aprobado' para definir 'aprobado' o 'suspenso'
    paso 4 : Si bonus=True, contar cu√°ntas notas individuales est√°n >= nota_aprobado
    paso 5 : Devolver la media y el estado (y el conteo de aprobados si bonus=True)
    """

    # 1Ô∏è‚É£ Verificar lista vac√≠a
    if not numeros:
        return "La lista est√° vac√≠a"

    # 2Ô∏è‚É£ Calcular la media
    suma_total = sum(numeros)
    media = suma_total / len(numeros)

    # 3Ô∏è‚É£ Determinar el estado seg√∫n la media
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"

    # 4Ô∏è‚É£ BONUS: contar cu√°ntas notas individuales est√°n aprobadas
    if bonus:
        aprobados = 0
        for nota in numeros:
            if nota >= nota_aprobado:
                aprobados += 1
        # 5Ô∏è‚É£ Devolver media, estado y n√∫mero de aprobados
        return media, estado, aprobados

    # 5Ô∏è‚É£ (sin bonus) Devolver media y estado
    return media, estado
```

```python
# ejemplos de uso
notas = [4, 7, 8, 5, 3]

# declaramos las variables para capturar los valores retornados
media, estado, aprobados = calcular_media_estado(notas, bonus=True)

print(f"Media: {media}")
print(f"Estado: {estado}")
print(f"examenes aprobados: {aprobados}")
```
