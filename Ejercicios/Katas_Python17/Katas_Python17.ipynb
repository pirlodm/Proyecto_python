## üìò Katas de Python 17 - Unir D√≠gitos (Reduce)

Crea una funci√≥n que tome una lista de d√≠gitos y devuelva el n√∫mero correspondiente. Por ejemplo, (5,7,2) 
corresponde al n√∫mero quinientos setenta y dos (572). Usa la funci√≥n 
reduce()

```python
# Katas de Python 17 - Unir D√≠gitos (Reduce)

# 0Ô∏è‚É£ Paso previo obligatorio: Importar reduce
from functools import reduce

def unir_lista_digitos(lista_digitos):
    """
    Toma una lista de d√≠gitos [5, 7, 2] y la reduce a un solo n√∫mero entero 572.
    
    paso 1: la funci√≥n auxiliar define C√ìMO se unen dos n√∫meros
            (Multiplicar el acumulado por 10 y sumar el nuevo d√≠gito)
    paso 2: Aplicar reduce() usando esa funci√≥n sobre la lista
    paso 3: Devolver el resultado final
    """
    # 1Ô∏è‚É£ dreamos la funcion auxiliar , logica matematica
    def unir(acumulado, nuevo):
        return (acumulado * 10) + nuevo
    
    # 2Ô∏è‚É£ Aplicar Reduce
    numero_final = reduce(unir, lista_digitos)

    # 3Ô∏è‚É£ Retorno
    return numero_final
```

```python
# ejemplo de uso 
# 1. a√±adimos los datos
digitos = [5,7,2]

# 2. llamamos a la funcion unir_lista_digitos
resultado = unir_lista_digitos(digitos)

# 3 obtenemos los resultados
print(f"lista original: {digitos}")
print(f"Numeros unidos: {resultado}")
```

## üí° BONUS - Calculadora de Totales

```python
# BONUS - Calculadora de Totales

def calculadora_total(lista_productos_digitos):
    """
    Recorre una lista de precios 'rotos' (d√≠gitos sueltos), 
    reconstruye cada precio usando TU funci√≥n y calcula la suma total.
    
    paso 1: Inicializar una variable 'suma_total' en 0
    paso 2: Recorrer cada producto de la lista
    paso 3: Usar TU funci√≥n 'unir_lista_digitos' para obtener el precio real del producto
    paso 4: Sumar ese precio a 'suma_total'
    paso 5: Mostrar el total acumulado
    """
    # 1Ô∏è‚É£ iniciamos acumulador en 0
    suma_total = 0

    print("--- üõí PROCESANDO DATOS---")

    # 2Ô∏è‚É£ Bucle (Iteramos sobre el lote)
    for producto_digitos in lista_productos_digitos:

        # 3Ô∏è‚É£ usamos la funcion (unir_lista_digitos)
        precio_real = unir_lista_digitos(producto_digitos)

        print(f"Detectado precio de: {precio_real} Euros")

        # 4Ô∏è‚É£ Sumamos al total
        suma_total = suma_total + precio_real
    
    # 5Ô∏è‚É£ Reporte Final
    print(f"El precio total es : {suma_total} Euros")
```

```python
# ejemplo de uso 
# 1. a√±adimos los datos
carrito_compras = [
    [1, 5],
    [5],
    [7, 1],
    [1, 0]
]

# 2. ejecutamos la calculadora
calculadora_total(carrito_compras)
```
