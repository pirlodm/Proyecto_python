Crea una funci√≥n que calcule el cubo de un n√∫mero dado mediante una funci√≥n 
lambda


```python
# Katas de Python 21 - Calcular el Cubo (Lambda)

# paso 1 creamos la funcion lamba 
# recibe un numero y devuelve ese numero elevado a 3
calcular_cubo = lambda numero: numero ** 3
```

```python
# ejemplos de uso 

# datos crudos
dato = 6

# usamos lambda
resultado = calcular_cubo(dato)

# obtebemos resultados
print(f" el numero es: {dato}")
print(f" su cubo es : {resultado}")
```
