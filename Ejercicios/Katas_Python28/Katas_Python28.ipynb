## üìò Katas de Python 28 - El Primer Duplicado

Crea una funci√≥n que busque y devuelva el primer elemento duplicado en una lista dada

```python
# Katas de Python 28 - El Primer Duplicado

def encontrar_primer_duplicado(lista_datos):
    """
    Recorre una lista y devuelve el primer elemento que se repite.
    Si no hay duplicados, devuelve None (nada).
    
    paso 1: Crear una "libreta" vac√≠a (set) para apuntar lo visto
    paso 2: Recorrer cada dato de la lista original
    paso 3: pregunta si ya esta ese dato en la libreta:
            - SI: ¬°Lo encontr√©! Devuelvo el dato y termino.
            - NO: Lo apunto en la libreta y sigo con el siguiente.
    """

    # paso 1: creamos libreta vacia
    vistos = set()

    # paso 2: recorremos la lista original 
    for dato in lista_datos:
        # paso 3: preguntamos si ya esta en la libreta
        if dato in vistos:
            # si ya esta en visto, es que salio antes
            return dato
        # si no estaba , lo guardamos para la proxima
        vistos.add(dato)

    # si  termina el bucle y no devuelve nada , es que todos eran unicos
    return None
```

```python
# ejemplos de uso

# ejercicio con un duplicado
mis_numeros = [10, 20, 30, 20, 50]
resultado = encontrar_primer_duplicado(mis_numeros)

print(f"Lista: {mis_numeros}")
print(f"El primer duplicado es: {resultado}")
```

## üí° BONUS - Integridad de Primary Keys

```python
# BONUS - Integridad de Primary Keys

def integridad_base_datos(lista_ids):
    """
    Verifica que no existan IDs repetidos en una tabla.
    
    paso 1: Usar la funci√≥n 'encontrar_primer_duplicado'
    paso 2: Verificar el resultado
            - Si hay resultado: ALERTA ROJA (Base de datos corrupta).
            - Si es None: VERDE (Datos limpios).
    """
    print("--- üîç AUDITOR√çA DE INTEGRIDAD ---")

    # paso 1: usamos la funcion
    error_encontrado = encontrar_primer_duplicado(lista_ids)

    # paso 2: verificamos que resultado nos da 
    if error_encontrado:
        print(f"‚ùå ERROR CR√çTICO: Violaci√≥n de ids.")
        print(f"   El ID '{error_encontrado}' est√° duplicado.")
    
    else:
        print("‚úÖ INTEGRIDAD OK: Todos los IDs son √∫nicos.")

```

```python
# ejemplos de uso 

# ejercicio 1 
ids_transacciones = [101, 102, 103, 102, 104]

integridad_base_datos(ids_transacciones)
```

```python
# ejercicio 2
ids_correctos = [101, 102, 103, 104, 105]

integridad_base_datos(ids_correctos)
```
