## üìò Katas de Python 20 - Limpieza de Tipos de Datos

 Para una lista con elementos tipo integer y string obt√©n una nueva lista s√≥lo con los valores int. Usa la funci√≥n 
filter()


```python
# Katas de Python 10 - Limpieza de Tipos de Datos

def filtrar_solo_enteros(lista_mixta):
    """
    Recibe una lista con n√∫meros y textos, y devuelve solo los n√∫meros enteros.
    
    paso 1: Definir funci√≥n auxiliar que verifique el tipo de dato
    paso 2: La condici√≥n es: isinstance(x, int) -> ¬øEs entero?
    paso 3: Usar filter() sobre la lista sucia
    paso 4: Devolver la lista limpia
    """

    # paso 1: definimos funcion auxiliar 
    def numero_entero(dato):
        # devuelve true solo si es "int"
        return isinstance(dato, int)
    
    # paso 2 y 3: aplicamos el filtro
    resultado = filter(numero_entero, lista_mixta)

    # paso 4: retornamos a lista
    return list(resultado)
```

```python
# ejemplos de uso 

# datos crudos
datos_sucios = [100, "Error", 200, "Sin Dato", 50, "N/A"]

# aplicamos la funcion
datos_limpios = filtrar_solo_enteros(datos_sucios)

# obtenemos los resultados 
print(f"datos Sucios : {datos_sucios}")
print(f"datos Limpios: {datos_limpios}")



```

## üí° BONUS - Bonus de Calidad de los Datos

```python
def medir_calidad_datos(lista_mixta):
    """
    Mide la calidad de los datos.
    
    paso 1: Obtener el total de datos brutos 
    paso 2: Usar la funci√≥n anterior para obtener solo los datos v√°lidos
    paso 3: Calcular las m√©tricas y porcentajes (C√°lculo)
    paso 4: Mostrar la evidencia y el resultado final (Salida)
    """
    
    # paso 1: entrada: medimos el total
    total_datos = len(lista_mixta)
    
    # paso 2: usamos lafunci√≥n de limpieza
    datos_limpios = filtrar_solo_enteros(lista_mixta)
    
    # paso 3: calculamos los porcentajes
    cantidad_utiles = len(datos_limpios)
    
    if total_datos > 0:
        calidad = (cantidad_utiles / total_datos) * 100
    else:
        calidad = 0
        
    # paso 4: mostramos los resultados

    print(f"---REPORTE DE CALIDAD---")
    print(f"Evidencia (Datos rescatados): {datos_limpios}") # ¬°Aqu√≠ est√° el print que pediste!
    print(f"Estad√≠stica: {cantidad_utiles} de {total_datos} son v√°lidos.")
    print(f"Puntuaci√≥n Final: {calidad:.1f}%")
```

```python
# ejemplos de uso 

# Una lista muy sucia (Mucho texto, pocos n√∫meros)
datos_archivo = ["Error", 10, "N/A", "Ups", 20, 50, "Basura", 30]

# Medimos la calidad
medir_calidad_datos(datos_archivo)
```
