##  Katas de Python 23 - Unir Palabras (Concatenar)

Concatena una lista de palabras.Usa la funci贸n 
reduce() .

```python
# Katas de Python 23 - Unir Palabras (Concatenar)

# importamos la funcion reduce 
from functools import reduce

def unir_frase_completa(lista_palabras):
    """
    Toma una lista de palabras sueltas y las une en una sola frase
    separ谩ndolas con espacios.
    
    paso 1: Definir una funci贸n auxiliar que sepa unir DOS palabras
    paso 2: Usar reduce() para aplicar eso a toda la lista
    """

    # paso 1: funcion auxiliar que une palabras y espacios
    def pegar_con_espacios(acumulado, nueva_palabra):
        return acumulado + " " + nueva_palabra
    
    # paso 2: usamos redurce para que itere la funcion en toda la lista 
    frase_final = reduce(pegar_con_espacios, lista_palabras)

    return frase_final
```

```python
# ejemplos de uso 

palabras = ["detras", "del", "ultimo", "no", "va", "nadie"]

resultado = unir_frase_completa(palabras)

print(f"Lista original: {palabras}")
print(f"Frase unida   : '{resultado}'")
```

##  BONUS - Generador de Consultas SQL

```python
# BONUS - Generador de Consultas SQL

def construir_query_sql(fragmentos_query):
    """
    Recibe las partes de una instrucci贸n SQL y construye la frase completa
    para enviarla a la base de datos.
    
    paso 1: Usar la funci贸n 'unir_frase_completa' para crear el string
    paso 2: Mostrar la consulta lista para ejecutar
    """
    print("----------- CONSTRUCTOR SQL -----------")

    # paso 1: usamos nuestra funcion anterior
    query_final = unir_frase_completa(fragmentos_query)

    # paso 2: mostramos el resultado listo
    print(f"fragmentos: {fragmentos_query}")
    print(f"ejecutando query: \n >> {query_final};")
```

```python
# ejemplo de uso 
partes_sql = ["SELECT", "nombre, email", "FROM", "clientes", "WHERE", "edad > 18"]
construir_query_sql(partes_sql)
```
