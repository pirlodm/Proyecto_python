## üìò Katas de Python 27 - Calcular el Promedio

Crea una funci√≥n que calcule el promedio de una lista de n√∫meros.

```python
# Katas de Python 27 - Calcular el Promedio

def calcular_promedio_seguro(lista_numeros):
    """
    Calcula el promedio, pero primero verifica si hay datos.
    Evita el error de 'dividir por cero'.
    
    paso 1: Contar cu√°ntos datos hay
    paso 2: Si la cantidad es 0 -> Devolver 0 inmediatamente (Protecci√≥n).
    paso 3: Si hay datos -> Calcular y devolver el promedio normal.
    """
    
    # Paso 1: Contamos 
    cantidad = len(lista_numeros)
    
    # Pase 2: zona de proteccion, si no hay nada no podemos repatir
    if cantidad == 0:
        return 0  # Devolvemos 0 y la funci√≥n termina aqu√≠.
    
    # Paso 3: calculamos la suma y el promedio
    # Si el c√≥digo llega aqu√≠, es que cantidad > 0, as√≠ que es seguro dividir.
    suma = sum(lista_numeros)
    promedio = suma / cantidad
    
    return promedio
```

```python
# ejemplos de uso 

# ejemplo 1
notas_clase = [10, 20, 30, 25, 15, 20]
print(f"Lista llena : {calcular_promedio_seguro(notas_clase)}")

# ejemplo 2
lista_vacia = []
print(f"Lista vac√≠a : {calcular_promedio_seguro(lista_vacia)}")
```

## üí° BONUS - El KPI de Ventas

```python
# BONUS - El KPI de Ventas

def evaluar_rendimiento_mes(lista_ventas_diarias, meta_objetivo):
    """
    Calcula si el promedio de ventas del mes super√≥ el objetivo de la empresa.
    
    paso 1: Usar la funci√≥n 'calcular_promedio' para obtener la media real
    paso 2: Comparar el promedio contra la Meta (Target)
    paso 3: Mostrar si se cumpli√≥ o no (KPI Status)
    """
    
    print(f"--- üìä REPORTE MENSUAL (Meta: {meta_objetivo}) ---")
    
    # paso 1: calculamos usando la funcion anterior
    promedio_real = calcular_promedio_seguro(lista_ventas_diarias)
    
    print(f"Ventas diarias: {lista_ventas_diarias}")
    print(f"Promedio conseguido: {promedio_real:.2f}") 
    
    # Paso 2 y 3: comparaamos el promedio
    if promedio_real >= meta_objetivo:
        print("‚úÖ RESULTADO: ¬°Objetivo Cumplido! Buen trabajo.")
    else:
        # Calculamos cu√°nto falt√≥ para dar feedback √∫til
        diferencia = meta_objetivo - promedio_real
        print(f"‚ùå RESULTADO: Objetivo no alcanzado. Faltaron {diferencia:.2f} puntos de media.")

```

```python
# ejemplo de uso

target_empresa = 50

# Ejercicio 1
ventas_equipo_A = [60, 70, 40] 
evaluar_rendimiento_mes(ventas_equipo_A, target_empresa)

# Ejercicio 2
ventas_equipo_B = [20, 30, 40]
evaluar_rendimiento_mes(ventas_equipo_B, target_empresa)
```
