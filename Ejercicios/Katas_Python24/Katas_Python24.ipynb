##  Katas de Python 24 - Diferencia Total (Resta en cadena)

Calcula la diferencia total en los valores de una lista. Usa la funci贸n 
reduce() .

```python
# Katas de Python 24 - Diferencia Total (Resta en cadena)

# importamos la funcion reduce
from functools import reduce

def calcular_resta_total(lista_numeros):
    """
    Toma una lista y resta los valores secuencialmente.
    El primer n煤mero es el total inicial, los demas le restan.
    
    paso 1: Definir funci贸n auxiliar (a - b)
    paso 2: Usar reduce() para aplicar la resta a toda la lista
    """
    # paso 1: definimos la funcion auxiliar restar
    def restar(acumulado, nuevo_valor):
        return acumulado - nuevo_valor
    
    # paso 2: usamos reduce 
    resultado = reduce(restar, lista_numeros)

    return resultado
```

```python
# ejemplos de uso 

valores = [100, 20, 10, 5, 20]
resultado_final = calcular_resta_total(valores)

print(f"Valores: {valores}")
print(f"El Resultado final: {resultado_final}")
```

```python
def gestionar_stock_real(stock_inicial, lista_de_ventas):
    """
    Recibe el stock inicial por un lado y la lista de ventas por otro.
    Los combina y calcula el resultado final.
    
    paso 1: Crear una lista 煤nica poniendo el stock al principio
    paso 2: Usar la funci贸n 'calcular_resta_total' con esa lista nueva
    paso 3: Mostrar el resultado
    """
    
    # paso 1:
    # Juntamos el dato suelto con la lista
    datos_combinados = [stock_inicial] + lista_de_ventas
    
    # paso 2: Usamos la funci贸n 'calcular_resta_total'
    stock_final = calcular_resta_total(datos_combinados)
    
    # 3锔 Reportemos el resultados
    print("---  GESTIN DE ALMACN ---")
    print(f"Stock Inicial : {stock_inicial}")
    print(f"Ventas del d铆a: {lista_de_ventas}")
    print(f"Stock Final   : {stock_final}")

```

```python
# ejemplos de uso

# Ahora los datos tienen mucho m谩s sentido l贸gico:
mi_stock = 150
mis_ventas = [10, 5, 20, 30, 50]

# Ejecutamos
gestionar_stock_real(mi_stock, mis_ventas)
```
