## üìò Katas de Python 10 - Excepciones Personalizadas (Promedio)

Escribe una funci√≥n que reciba una lista de n√∫meros y calcule su promedio. Si la lista est√° vac√≠a, lanza una 
excepci√≥n personalizada y maneja el error adecuadamente.

```python
# Katas de Python 10 - Excepciones Personalizadas (Promedio)

# Paso previo: Definir nuestra propia Excepci√≥n

class listaVaciaError(Exception):
    """Excepci√≥n personalizada para indicar que la lista est√° vac√≠a."""
    pass

def calcular_promedio(lista_numeros):
    """
    Calcula el promedio de una lista. Si est√° vac√≠a, lanza un error personalizado.
    
    paso 1: Verificar si la lista tiene elementos (len > 0)
    paso 2: Si est√° vac√≠a, LANZAR (raise) la excepci√≥n personalizada
    paso 3: Si tiene datos, calcular la suma dividida por la cantidad
    paso 4: Devolver el resultado
    """

    # # 1Ô∏è‚É£ y 2Ô∏è‚É£ Validaci√≥n
    if len(lista_numeros) == 0:
        # aqui detenemos la ejecuci√≥n y lanzamos el error
        raise listaVaciaError("No se puede calcular el promedio: La lista no tiene datos.")
    # 3Ô∏è‚É£ C√°lculo (suma / cantidad)
    promedio = sum(lista_numeros) / len(lista_numeros)

    # 4Ô∏è‚É£ Devoluci√≥n
    return promedio
```

```python
# ejemplos de uso para el error personalizado

datos_error = [] # lista vacia para probrar el error

try:
    resultado = calcular_promedio(datos_error)
    print(f"El promedio es: {resultado}")

except listaVaciaError as error_detectado:
    print(f"Error: {error_detectado}")
```

```python
# ejemplos de uso si error

datos_validos = [10, 20, 30, 40, 50] # lista con datos

try:
    resultado = calcular_promedio(datos_validos)
    print(f"El promedio es: {resultado}")

except listaVaciaError as error_detectado:
    print(f"Error: {error_detectado}")

```

## üí° BONUS - obtener la media y luego hacer una comparaci√≥n simple

```python
# obtener la media y luego hace una comparaci√≥n simple.

def comparar_con_media(lista_datos, mi_dato):
    """
    Compara un valor individual contra el promedio del grupo.
    Usa la funci√≥n 'calcular_promedio' del ejercicio anterior.
    
    paso 1: Intentar calcular el promedio del grupo
    paso 2: Si la lista est√° vac√≠a, capturar el error y avisar
    paso 3: Si hay promedio, comparar si 'mi_dato' es mayor o menor
    paso 4: Devolver el resultado de la comparaci√≥n
    """

    try:
        # 1Ô∏è‚É£ reutilizmaos la funcion del ejercicio anterior
        promedio = calcular_promedio(lista_datos)

        # 3Ô∏è‚É£ Comparaci√≥n
        if mi_dato > promedio:
            return f"tu dato {mi_dato} esta por encima del promedio {promedio}"
        else:
            return f"tu dato {mi_dato} esta por debajo del promedio {promedio}"
        
    except listaVaciaError:
        # 2Ô∏è‚É£ Capturamos el error y avisamos
        return "No se puede comparar: La lista de datos est√° vac√≠a."
```

```python
# ejemplo de uso

# caso A: datos normales 
ventas_tienda = [150, 200, 250, 300, 350]
mi_venta = 275
print(comparar_con_media(ventas_tienda, mi_venta))

# caso B: lista vacia
ventas_tienda_vacia = []
mi_venta = 275
print(comparar_con_media(ventas_tienda_vacia, mi_venta))
```
