## üìò Katas de Python 11 - Validaci√≥n de Edad (Rango y Tipo)

Escribe un programa que pida al usuario que introduzca su edad. Si el usuario ingresa un valor no num√©rico o un 
valor fuera del rango esperado (por ejemplo, menor que 0 o mayor que 120), maneja las excepciones 
adecuadamente

```python
# Katas de Python 11 - Validaci√≥n de Edad (Rango y Tipo)

def pedir_edad_usuario():
    """
    Solicita la edad y valida que sea un n√∫mero l√≥gico (0-120).
    
    paso 1: Pedir al usuario que introduzca su edad
    paso 2: Intentar convertir la entrada a n√∫mero entero (int)
    paso 3: Verificar si el n√∫mero est√° dentro del rango l√≥gico (0 a 120)
    paso 4: Si es v√°lido, devolver la edad
    paso 5: Si est√° fuera de rango, avisar del error
    paso 6: Si no es un n√∫mero (ValueError), avisar del error de formato
    """
    
    print("--- Verificaci√≥n de Edad ---")

    try:
        # 1Ô∏è‚É£ y 2Ô∏è‚É£ Entrada y conversi√≥n
        entrada = int(input("Por favor, introduce tu edad: "))
        edad = int(entrada)

        # 3Ô∏è‚É£ Verificaci√≥n del rango l√≥gico
        if 0 <= edad <= 120:
            # 4Ô∏è‚É£ Edad v√°lida
            print(f"‚úÖ Edad registrada correctamente: {edad}")
            return edad
        else:
            # 5Ô∏è‚É£ Edad fuera de rango
            print("‚ùå Error: La edad debe estar entre 0 y 120 a√±os.")
            return None
    
    except ValueError:
        # 6Ô∏è‚É£ Error de Tipo (Escribieron letras)
        print("‚ùå Error: Formato incorrecto. Debes introducir un n√∫mero entero.")
        return None
```

```python
# ejemplo de uso 

edad_valida = pedir_edad_usuario()
```

## üí° BONUS - Segmentaci√≥n de rango de edades


```python
# segmentacion de rango de edades

def clasificar_edad(edad):
    """
    Toma una edad num√©rica y devuelve la categor√≠a correspondiente.
    Complementa la funci√≥n anterior clasificando el dato obtenido.
    
    paso 1: Verificar que recibimos un dato v√°lido (no None)
    paso 2: Si es menor de 18 -> Categor√≠a 'Menor'
    paso 3: Si es entre 18 y 65 -> Categor√≠a 'Adulto Activo'
    paso 4: Si es mayor de 65 -> Categor√≠a 'Senior'
    paso 5: Mostrar la clasificaci√≥n
    """
    # 1Ô∏è‚É£ Protecci√≥n (Por si la funci√≥n anterior devolvi√≥ None
    if edad is None:
        print("no se puede clasificar: edad no valida")
        return
    
    # 2Ô∏è‚É£, 3Ô∏è‚É£ y 4Ô∏è‚É£ Segmentaci√≥n de la clasificaci√≥n
    if edad < 18:
        segmento = "menor de edada"
    elif 18 <= edad <= 65:
        segmento = "adulto activo"
    else:
        segmento = "senior"

    # 5Ô∏è‚É£ resultado de la clasificaci√≥n
    print(f"Tu clasificaci√≥n de edad es: {segmento}.")
    
```

```python
# ejemplo de uso conjunto 

# 1. Obtenemos el dato limpio con TU funci√≥n
mi_edad = pedir_edad_usuario()

# 2. Clasificamos el dato con el BONUS
clasificar_edad(mi_edad)
```
