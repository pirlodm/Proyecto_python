## üìò Katas de Python 09 - filtrar mascostas prohibidas

 Escribe una funci√≥n que tome una lista de nombres de mascotas como par√°metro y devuelva una nueva lista 
excluyendo ciertas mascotas prohibidas en Espa√±a. La lista de mascotas a excluir es ["Mapache", "Tigre", 
"Serpiente Pit√≥n", "Cocodrilo", "Oso"].Usa la funci√≥n 
filter()

```python
# Katas de Python 09 - filtrar mascostas prohibidas

def filtrar_mascotas(lista_mascotas):
    """
    Filtra mascotas prohibidas ignorando may√∫sculas/min√∫sculas sin usar lambda.
    
    paso 1: Definir la lista de prohibidas normalizada (todo en min√∫sculas)
    paso 2: Crear una funci√≥n interna que decida si una mascota es v√°lida
    paso 3: Aplicar filter() usando esa funci√≥n interna
    paso 4: Convertir el resultado a lista y devolverlo
    """

    # 1Ô∏è‚É£ Lista de referencia en min√∫sculas
    prohibidas = ["mapache", "tigre", "cocodrilo","oso", "serpiente"]

    # 2Ô∏è‚É£ Funci√≥n "Juez": Decide qui√©n pasa y qui√©n no
    def es_mascota_valida(mascota):
        # convertir a min√∫sculas para comparar
        nombre_en_minuscula = mascota.lower()

        # devuelve true si NO  esta en la lista de prohibidas
        return nombre_en_minuscula not in prohibidas
    
    # 3Ô∏è‚É£ Filtrar usando la funci√≥n juez
    resultado = filter(es_mascota_valida, lista_mascotas)

    # 4Ô∏è‚É£ Convertir a lista y devolver
    return list(resultado)
```

```python
# ejemplos de uso 

lista = ["Perro", "tigre", "Gato", "OSO", "Mapache", "Pez","serpiente","oso"]

# llamamos a la funcion filtrar_mascotas prohibidas
lista_filtrada = filtrar_mascotas(lista)

print("--- lista completa ---")
print(f"lista original: {lista}")
print("--- Resultado del Filtrado ---")
print(f"lista filtrada: {lista_filtrada}")
```

## üí° BONUS - Pre-procesamiento (Limpiar Duplicados)

```python
# Pre-procesamiento (Limpiar Duplicados)

def eliminar_duplicados(lista_sucia):
    """
    Recibe una lista con datos repetidos y devuelve una lista con valores √∫nicos.
    Mantiene el orden original de aparici√≥n.
    
    paso 1: Crear una lista vac√≠a para guardar los datos √∫nicos
    paso 2: Recorrer cada elemento de la lista sucia
    paso 3: Preguntar: ¬øEste elemento ya est√° en mi lista de √∫nicos?
    paso 4: Si NO est√° -> Lo agrego (es nuevo)
    paso 5: Si S√ç est√° -> Lo ignoro (es repetido)
    paso 6: Devolver la lista limpia
    """
    # 1Ô∏è‚É£ Lista vac√≠a para los √∫nicos
    lista_unicos = []

    # 2Ô∏è‚É£ Recorrer cada elemento
    for element in lista_sucia:

        # 3Ô∏è‚É£ Preguntar si ya est√° en la lista de √∫nicos
        if element not in lista_unicos:
            # 4Ô∏è‚É£ Si NO est√°, lo agrego
            lista_unicos.append(element)
        # 5Ô∏è‚É£ Si S√ç est√°, lo ignoro (no hago nada)
    
    # 6Ô∏è‚É£ Devolver la lista limpia
    return lista_unicos
```

```python
# usar las dos funciones juntas

# 1. Datos de entrada (Muy sucios: repetidos y prohibidos)
datos_entrada = ["Perro", "Tigre", "Gato", "Perro", "Perro", "Tigre", "Oso"]
print(f"1. Entrada       : {datos_entrada}")

# 2. PRIMER PASO: Limpiar duplicados 
datos_sin_duplicados = eliminar_duplicados(datos_entrada)
print(f"2. Sin Duplicados: {datos_sin_duplicados}")

# 3. SEGUNDO PASO: Filtrar prohibidos 
# (Usando la funci√≥n filtrar)
datos_finales = filtrar_mascotas(datos_sin_duplicados)
print(f"3. Resultado Final: {datos_finales}")
```
