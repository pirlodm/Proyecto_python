##  Katas de Python 18 - Filtrado de Estructuras de Datos (JSON)

Escribe un programa en Python que cree una lista de diccionarios que contenga informaci贸n de estudiantes 
(nombre, edad, calificaci贸n) y use la funci贸n filter para extraer a los estudiantes con una calificaci贸n mayor o igual a 
90. Usa la funci贸n 
filter()

```python
# Katas de Python 18 - Filtrado de Estructuras de Datos (JSON)

def filtrar_mejores_estudiantes(lista_clase):
    """
    Recibe una lista de diccionarios (estudiantes) y devuelve solo los que 
    tienen una calificaci贸n de 90 o superior.
    
    paso 1: Definir la funci贸n que inspecciona el diccionario
    paso 2: La condici贸n es: estudiante['calificacion'] >= 90
    paso 3: Usar filter() para aplicar esa condici贸n a toda la lista
    paso 4: Devolver la lista filtrada
    """
    # 1锔 Funci贸n Auxiliar que inspecciona el diccionario
    def es_excelente(estudiante):
        #accedemos al valor por su calificacion
        nota = estudiante["calificacion"]
        # devolvemos si cumple el siguiente requisito
        return nota >= 90
    
    # 2锔 y 3锔 Aplicar Filtro
    resultado = filter(es_excelente, lista_clase)

    # 4锔 Retorno
    return list(resultado)
```

```python
# ejemplo de uso 

# 1. a帽adimos los datos
datos_clase = [
    {'nombre': 'Ana',    'edad': 20, 'calificacion': 85},
    {'nombre': 'Beto',   'edad': 22, 'calificacion': 95}, 
    {'nombre': 'Carla',  'edad': 19, 'calificacion': 90}, 
    {'nombre': 'Daniel', 'edad': 21, 'calificacion': 70},
    {'nombre': 'Elena',  'edad': 23, 'calificacion': 98},
    {'nombre': 'Ines',  'edad': 30, 'calificacion': 70}
]

# 2. llamamos a la funcion creada 
alumnos_top = filtrar_mejores_estudiantes(datos_clase)

# 3. obtenemos los resultados
for alumno in alumnos_top:
    print(f"Estudiante: {alumno['nombre']} \t Nota: {alumno['calificacion']}")
```

##  BONUS - comparativa con la media

```python
def comparar_con_media_real(lista_toda_la_clase, lista_filtrada_top):
    """
    Calcula la media de toda la clase y compara a los mejores alumnos contra ella.
    
    paso 1: Calcular la suma de TODAS las calificaciones de la clase
    paso 2: Dividir la suma entre el total de alumnos para sacar la MEDIA REAL
    paso 3: Recorrer solo a los alumnos de la lista filtrada (Los Mejores)
    paso 4: Restar su nota menos la media real para ver su "ventaja"
    paso 5: Mostrar el reporte
    """
    
    # --- PARTE 1: CALCULAR LA MEDIA REAL (Estad铆stica) ---
    
    suma_total = 0
    
    # 1锔 Sumamos las notas de todo el mundo (Aprobados y suspensos)
    for alumno in lista_toda_la_clase:
        suma_total = suma_total + alumno['calificacion']
        
    # 2锔 Calculamos el promedio
    media_real = suma_total / len(lista_toda_la_clase)
    
    print(f"Media del grupo: {media_real:.1f}")
    
    # --- PARTE 2: COMPARAR A LOS MEJORES (Tu filtro) ---
    
    # 3锔 Bucle solo sobre los cracks
    for alumno_top in lista_filtrada_top:
        
        nombre = alumno_top['nombre']
        nota = alumno_top['calificacion']
        
        # 4锔 C谩lculo de la ventaja (Benchmarking)
        # Si sac贸 95 y la media es 73.7, la ventaja es 21.3
        ventaja = nota - media_real
        
        # 5锔 Reporte
        print(f"Estudiante: {nombre} \t Nota: {nota} (+{ventaja:.1f} puntos sobre la media de la clase)")
```

```python
# ejemplos de uso 

#1 datos crudos
datos_clase = [
    {'nombre': 'Ana',    'edad': 20, 'calificacion': 85},
    {'nombre': 'Beto',   'edad': 22, 'calificacion': 95}, 
    {'nombre': 'Carla',  'edad': 19, 'calificacion': 90}, 
    {'nombre': 'Daniel', 'edad': 21, 'calificacion': 70},
    {'nombre': 'Elena',  'edad': 23, 'calificacion': 98},
    {'nombre': 'Ines',  'edad': 30, 'calificacion': 70}
]

#2 usamos la funcion anterior filtrar_mejores_estudiantes
los_mejores = filtrar_mejores_estudiantes(datos_clase)

#3 utilizamos el bonus solo a los que pasaron el filtro
comparar_con_media_real(datos_clase, los_mejores)
```
