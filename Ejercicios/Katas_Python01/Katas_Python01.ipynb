## üìò Katas de Python 01 - Frecuencia de letras

Escribe una funci√≥n que reciba una cadena de texto como par√°metro y devuelva un diccionario con las frecuencias  
de cada letra en la cadena. Los espacios no deben ser considerados.


```python
# Katas de Python 01 - Frecuencia de letras


def frecuencia_letras(texto):
    """
    Devuelve un diccionario con la frecuencia de cada letra en 'texto'

    paso 1 : Eliminar los espacios en blanco del texto
    paso 2 : Crear un diccionario vac√≠o para almacenar la frecuencia de las letras
    paso 3 : Recorrer cada letra en el texto
        - Si la letra ya est√° en 'freq', sumamos 1.
        - Si no, la agregamos con valor 1.
    paso 4: Devolver el diccionario 'freq' con todas las frecuencias.
    """
    texto = texto.replace(" ", "") # 1Ô∏è‚É£ Limpiar el texto
    frequencia = {} # 2Ô∏è‚É£ Inicializar diccionario

    for letra in texto:
        frequencia[letra] = frequencia.get(letra,0) +1 # 3Ô∏è‚É£ Recorrer letras y contar
    return frequencia # 4Ô∏è‚É£ Devolver el resultado
```

```python
# ejemplo de uso

cadena = "Estoy esperando"
resultado = frecuencia_letras(cadena)
print(resultado)
```

```python
# Tests autom√°ticos para verificar que la funci√≥n funciona correctamente

# Caso 1: texto simple
assert frecuencia_letras("hola") == {'h':1, 'o':1, 'l':1, 'a':1}

# Caso 2: texto con espacios
assert frecuencia_letras("hola mundo") == {
    'h':1, 'o':2, 'l':1, 'a':1, 'm':1, 'u':1, 'n':1, 'd':1
}
print("‚úÖ Todos los tests han pasado correctamente")
```

## üí° BONUS minusculas
Vamos a crear la misma version , pero que tenga en cuenta que mayusculas y minisculas sean iguales

```python
# funci√≥n que no distinga entre may√∫sculas y min√∫sculas

def frecuencia_letras_minus(texto):
    """
    Devuelve un diccionario con la frecuencia de cada letra en 'texto'

    paso 1 : Convertir el texto a min√∫sculas para unificar letras
    paso 2 : Eliminar los espacios en blanco del texto
    paso 3 : Crear un diccionario vac√≠o para almacenar la frecuencia de las letras
    paso 4 : Recorrer cada letra en el texto
        - Si la letra ya est√° en 'freq', sumamos 1.
        - Si no, la agregamos con valor 1.
    paso 5: Devolver el diccionario 'freq' con todas las frecuencias.
    """
    texto = texto.lower() # 1Ô∏è‚É£ Convertir a min√∫sculas
    texto = texto.replace(" ", "") # 2Ô∏è‚É£ Limpiar espacios
    frequencia = {} # 3Ô∏è‚É£ Inicializar diccionario

    for letra in texto:
        frequencia[letra] = frequencia.get(letra,0) +1 # 4Ô∏è‚É£ Recorrer letras y contar
    return frequencia # 5Ô∏è‚É£ Devolver resultado
```

```python
# ejemplo de uso

cadena = "Estoy esperando"
resultado = frecuencia_letras_minus(cadena)
print(resultado)
```
