Crea una funci贸n lambda que  sume 3 a cada n煤mero de una lista dada.

```python
# Katas de Python 15 - Transformaci贸n con Lambda

def sumar_tres_lote(lista_numeros):
    """
    Suma 3 a cada n煤mero de la lista usando una funci贸n lambda.
    
    paso 1: Definir la operaci贸n lambda: lambda x: x + 3
    paso 2: Aplicar map() para ejecutar esa lambda sobre cada n煤mero
    paso 3: Convertir el resultado a lista y devolver
    """
    # 1锔 y 2锔 definimos la operacion y aplicamos map () para iterar
    resultado = map(lambda x: x + 3, lista_numeros)
    # 3锔 Retorno
    return list(resultado)
```

```python
# ejemplo de uso 

mis_numeros = [10,20,30,40,50]
nuevos_numeros = sumar_tres_lote(mis_numeros)

print(f"lista original: {mis_numeros}")
print(f"lista mas 3: {nuevos_numeros}")
```
