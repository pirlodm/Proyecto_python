## üìò Katas de Python 13 - Generador de Tuplas (May√∫scula/Min√∫scula)

Genera una funci√≥n la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en 
may√∫sculas y min√∫sculas. Las letras no pueden estar repetidas .Usa la funci√≥n 
map()

```python
# Katas de Python 13 - Generador de Tuplas (May√∫scula/Min√∫scula)

def generar_pares_letras(texto):
    """
    Recibe un texto, elimina duplicados y devuelve una lista de tuplas
    con el formato (MAY√öSCULA, min√∫scula).
    
    paso 1: Convertir el texto a un conjunto (set) para eliminar repetidos autom√°ticamente
    paso 2: Definir una funci√≥n auxiliar que reciba una letra y devuelva la tupla
    paso 3: Usar map() para aplicar esa funci√≥n a cada letra √∫nica
    paso 4: Convertir el resultado a una lista y devolverla
    """
    # 1Ô∏è‚É£ Limpieza de duplicados funcion set
    caracteres_unicos = set(texto)

    # 2Ô∏è‚É£ Funci√≥n auxiliar para crear la tupla
    def crear_tupla(letra):
        # devuelve un par (tupla) con la letra en may√∫scula y min√∫scula
        return (letra.upper(), letra.lower())
    
    # 3Ô∏è‚É£ Usar map para aplicar la funci√≥n a cada letra √∫nica
    resultado_map = map(crear_tupla, caracteres_unicos)

    # 4Ô∏è‚É£ retorno como lista 
    return list(resultado_map)
```

```python
# ejemplo de uso 

frase = "banana" 

resultado = generar_pares_letras(frase)
print(f"Texto original: {frase}")
print(f"Tuplas √∫nicas : {resultado}")
```

## üí° BONUS - Bonus: Exportar a formato CSV

```python
# BONUS: Exportar como CSV Simulado 

def exportar_csv_simulado(lista_tuplas):
    """
    Recorre la lista de tuplas y las muestra como si fuera un archivo CSV.
    Ideal para visualizar c√≥mo quedar√≠an los datos en Excel.
    
    paso 1: Imprimir los encabezados de las columnas
    paso 2: Recorrer cada tupla (desempaquetando may√∫scula y min√∫scula)
    paso 3: Imprimir los valores separados por una coma
    """
    
    print("--- VISTA PREVIA DEL ARCHIVO CSV ---")
    
    # 1Ô∏è‚É£ Encabezados (Header)
    print("Mayuscula,Minuscula")
    
    # 2Ô∏è‚É£ Bucle para imprimir filas
    for mayuscula, minuscula in lista_tuplas:
        
        # 3Ô∏è‚É£ Formato CSV: dato,dato
        print(f"{mayuscula},{minuscula}")
```

```python
# --- USO CONJUNTO ---

# 1. Tu funci√≥n crea los datos
datos = generar_pares_letras("Hola") 
# Genera: [('H','h'), ('O','o'), ('L','l'), ('A','a')] (orden aleatorio)

# 2. El Bonus los "exporta"
exportar_csv_simulado(datos)
```
