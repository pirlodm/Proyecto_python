##  Katas de Python 07 - Convertir tuplas a strings

Genera una funci贸n que convierta una lista de tuplas a una lista de strings. Usa la funci贸n 
map()

```python
# Katas de Python 07 - Convertir tuplas a strings


def tuplas_a_strings(lista_tuplas):
    """
    Convierte una lista de tuplas en una lista de strings usando map().

    paso 1: Crear una funci贸n interna que convierta una tupla a string
    paso 2: Usar map() para aplicar esa funci贸n a cada tupla de la lista
    paso 3: Convertir el resultado de map a una lista
    paso 4: Devolver la nueva lista de strings
    """

    # 1锔 Funci贸n para convertir una tupla en string

    def convertir(t):
        return str(t)

    # 2锔 Aplicar map() a toda la lista
    restultado = map(convertir,lista_tuplas)

    # 3锔 Convertir el resultado a lista
    resultado = list(restultado)

    # 4锔 Devolver la nueva lista
    return resultado
```

```python
# ejemplos de uso 

lista = [(1, 2), (3, 4), ("hola","david")]
print(tuplas_a_strings(lista))
```

```python
# test automaticos de prueba

assert tuplas_a_strings([(1, 2), (3, 4), (5, 6)]) == ['(1, 2)', '(3, 4)', '(5, 6)']
assert tuplas_a_strings([]) == []

print("Todos los tests pasaron correctamente.")
```

##  BONUS - validar si las tuplas tienen solo valores num茅ricos

```python
# Bonus Katas de Python 07 - Validaci贸n num茅rica de tuplas

def validar_tuplas_numericas(lista_tuplas):
    """
    Devuelve una lista de True/False indicando si cada tupla contiene 煤nicamente valores num茅ricos.

    paso 1 : Definir funci贸n interna que verifica si todos los elementos de la tupla son int o float
    paso 2 : Aplicar map() para recorrer todas las tuplas
    paso 3 : Convertir el resultado a lista
    paso 4 : Devolver lista de validaci贸n
    """
    
    # 1锔 Funci贸n interna para verificar si todos los elementos son n煤meros
    def es_numerica(tupla):
        return all(isinstance(x, (int, float)) for x in tupla)
    
    # 2锔 Aplicar map() a todas las tuplas
    resultado = map(es_numerica, lista_tuplas)
    
    # 3锔 Convertir a lista
    resultado = list(resultado)
    
    # 4锔 Devolver resultado
    return resultado
```

```python
datos = [(1, 2), ("hola", 3), (5.5, 7), ("A", "B")]

print("Validaci贸n num茅rica:", validar_tuplas_numericas(datos))
```
