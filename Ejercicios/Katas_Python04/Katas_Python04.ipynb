## üìò Katas de Python 04 - Diferencia entre listas usando map() 

Genera una funci√≥n que calcule la diferencia entre los valores de dos listas. Usa la funci√≥n map ()

```python
# Katas de Python 04 - Diferencia entre listas usando map()

def diferenciar_listas(lista1, lista2):
    """
    Devuelve una nueva lista con la diferencia de cada elemento de lista1 menos lista2.

    paso 1 : Definir una funci√≥n interna que calcule la diferencia entre dos n√∫meros
    paso 2 : Aplicar map() para recorrer ambas listas simult√°neamente
    paso 3 : Convertir el resultado a lista
    paso 4 : Devolver la lista de diferencias
    """

    # 1Ô∏è‚É£ Funci√≥n para calcular la diferencia
    def resta(x, y):
        return x - y
    
    # 2Ô∏è‚É£ Usar map() para aplicar la funci√≥n a ambas listas
    lista_diferencia = map(resta, lista1, lista2)

    # 3Ô∏è‚É£ Convertir a lista
    lista_diferencia = list(lista_diferencia)

    # 4Ô∏è‚É£ Devolver resultado

    return lista_diferencia
```

```python
# ejemplos de uso 

lista_a = [10, 20, 30, 40]
lista_b = [2,4,6,8]

resultado = diferenciar_listas(lista_a,lista_b)
print("lista 1:", lista_a)
print("lista 2:", lista_b)
print("Diferencia lista 1 - lista 2:", resultado)

```

```python
# tests atom√°ticos de prueba

assert diferenciar_listas([5, 10, 15], [2, 3, 5]) == [3, 7, 10]
assert diferenciar_listas([0, 0], [0, 0]) == [0, 0]
print("‚úÖ Todos los tests pasaron correctamente")
```
