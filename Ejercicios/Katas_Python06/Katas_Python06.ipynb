## üìò Katas de Python 06 - Factorial recursivo

 Escribe una funci√≥n que calcule el factorial de un n√∫mero de manera recursiva.

```python
# Katas de Python 06 - Factorial recursivo

def factorial(n):
    """
    Calcula el factorial de un n√∫mero de manera recursiva.

    paso 1 : Comprobar que n sea >= 0
    paso 2 : Caso base: si n == 0 o n == 1, devolver 1
    paso 3 : Caso recursivo: multiplicar n por factorial(n-1)
    paso 4 : Devolver el resultado
    """
    # 1Ô∏è‚É£ Comprobar que n sea v√°lido
    if n < 0:
        return "Error: El n√∫mero debe ser mayor o igual a 0."
    
    # 2Ô∏è‚É£ Caso base
    if n == 0 or n == 1:
        return 1
    
    # 3Ô∏è‚É£ Caso recursivo
    return n * factorial(n - 1)
```

```python
# ejemplos de uso

print(factorial(5))
print(factorial(0))
print(factorial(-3))
```
